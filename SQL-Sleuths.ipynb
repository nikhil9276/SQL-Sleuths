# Best Salesperson Based on Total Sales

```python
query = """
SELECT Salesperson, SUM(Sales) AS Total_Sales
FROM Sales
JOIN Salesperson ON Sales.EmployeeKey = Salesperson.EmployeeKey
GROUP BY Salesperson
ORDER BY Total_Sales DESC
LIMIT 5;
"""
df_best_salesperson = pd.read_sql_query(query, engine)
print(df_best_salesperson)
```

# Sales by Region

```python
query = """
SELECT Region, SUM(Sales) AS Total_Sales
FROM Sales
JOIN Region ON Sales.SalesTerritoryKey = Region.SalesTerritoryKey
GROUP BY Region
ORDER BY Total_Sales DESC;
"""
df_region_sales = pd.read_sql_query(query, engine)
print(df_region_sales)
```

# Most Sold Products (by Quantity)

```python
query = """
SELECT p.Product, SUM(s.Quantity) AS Total_Quantity
FROM Sales s
JOIN Product p ON s.ProductKey = p.ProductKey
GROUP BY p.Product
ORDER BY Total_Quantity DESC
LIMIT 10;
"""
df_top_products = pd.read_sql_query(query, engine)
print(df_top_products)
```

```python
import pandas as pd
from sqlalchemy import create_engine

# âœ… Read CSV files
product = pd.read_csv("CWP/Product.csv", delimiter="\t")
region = pd.read_csv("CWP/Region.csv", delimiter="\t")
reseller = pd.read_csv("CWP/Reseller.csv", delimiter="\t")
sales = pd.read_csv("CWP/Sales.csv", delimiter="\t")
salesperson = pd.read_csv("CWP/Salesperson.csv", delimiter="\t")
salesperson_region = pd.read_csv("CWP/SalespersonRegion.csv", delimiter="\t")
targets = pd.read_csv("CWP/Targets.csv", delimiter="\t")

# âœ… Data Cleaning & Type Conversion

## ðŸ“Œ Product Table
product.drop(columns=["Background Color Format", "Font Color Format"], inplace=True, errors="ignore")  
product["Standard Cost"] = pd.to_numeric(product["Standard Cost"], errors="coerce")  

## ðŸ“Œ Sales Table
sales["OrderDate"] = pd.to_datetime(sales["OrderDate"], errors="coerce")  
sales["Quantity"] = pd.to_numeric(sales["Quantity"], errors="coerce", downcast="integer")  
sales["Unit Price"] = pd.to_numeric(sales["Unit Price"].replace("[\$,]", "", regex=True), errors="coerce")  
sales["Sales"] = pd.to_numeric(sales["Sales"].replace("[\$,]", "", regex=True), errors="coerce")  
sales["Cost"] = pd.to_numeric(sales["Cost"].replace("[\$,]", "", regex=True), errors="coerce")  

## ðŸ“Œ Targets Table
targets["Target"] = pd.to_numeric(targets["Target"].replace("[\$,]", "", regex=True), errors="coerce")  
targets["TargetMonth"] = pd.to_datetime(targets["TargetMonth"], errors="coerce")  

# âœ… Verify Cleaned Data Before Inserting
print("âœ” Data cleaned successfully!")

# âœ… Establish Connection to MySQL
engine = create_engine("mysql+pymysql://root:n1k2i3@localhost/business_insights")
print("âœ” Connected to MySQL successfully!")

# âœ… Dictionary of table names and corresponding DataFrames
tables = {
    "Product": product,
    "Region": region,
    "Reseller": reseller,
    "Sales": sales,
    "Salesperson": salesperson,
    "SalespersonRegion": salesperson_region,
    "Targets": targets
}

# âœ… Insert Data into MySQL
for table_name, df in tables.items():
    try:
        df.to_sql(table_name, con=engine, if_exists="replace", index=False, chunksize=1000)
        print(f"âœ” Inserted {table_name} successfully!")
    except Exception as e:
        print(f"âŒ Error inserting {table_name}: {e}")

print("âœ… All data inserted into MySQL successfully!")
```
